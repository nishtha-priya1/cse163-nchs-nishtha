Your ouptut image should look something like the following:

````python
import imageio as iio
import matplotlib.pyplot as plt
import numpy as np

# Write your function here!
def edge_detect(image_file: str) -> None:
    # Define the kernel
    kernel = np.array([[-1, -1, -1], [-1, 8, -1], [-1, -1, -1]])
    # Apply the kernel
    image = iio.imread(image_file) 
    image = image.astype(np.float64)
    image = image[:, :, 0]
    height, width = np.shape(image)
    output = np.zeros((height, width), dtype=np.float32)
    

    for i in range(1, height - 1):
        for j in range(1, width-1):
            window = image[i-1:i+2, j-1:j+2]
            value = np.sum(window * kernel)
            output[i, j] = value

    output = np.abs(output)
    output = np.clip(output, 0, 255)

    binary = np.zeros_like(output, dtype=np.uint8)
    binary[output>127] = 255
    plt.imshow(binary, cmap= "gray")
    plt.show()
        
    
edge_detect("duck.jpg")
````
